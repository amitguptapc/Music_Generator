### Preprocessing all midi files

```python
notes = []
for file in glob.glob('./midi_songs/*.mid'):
    midi = converter.parse(file)
    print("Parsing : %s"%file.split('/')[-1])
    elements = midi.flat.notes
    for e in elements:
        if isinstance(e,note.Note):
            notes.append(str(e.pitch))
        elif isinstance(e,chord.Chord):
            notes.append('+'.join(str(n) for n in e.normalOrder))
```

```python
with open('notes.pkl','wb') as f:
    pickle.dump(notes,f)
```

```python
with open('notes.pkl','rb') as f:
    notes = pickle.load(f)
```

```python
n_vocab = len(set(notes))
print('Total Notes :',len(notes))
print('Unique Notes :',n_vocab)
```

### Prepare Sequential Data for LSTM

```python
seq_len = 100 # Length of data fot LSTM
classes = sorted(set(notes)) 
# map classes to integer
note2int = dict((e,num) for num,e in enumerate(classes))
```

```python
X = [] # input data for LSTM
Y = [] # output data for LSTM
for i in range(len(notes)-seq_len):
    inp = notes[i:i+seq_len]
    out = notes[i+seq_len]
    X.append([note2int[note] for note in inp])
    Y.append(note2int[out])
```

```python
X = np.reshape(X,(len(X),seq_len,1))
X.shape
```

```python
# Normalize the data
normalized_X = X/float(n_vocab)
# convert output to one hot vector
Y = np_utils.to_categorical(Y)
```
